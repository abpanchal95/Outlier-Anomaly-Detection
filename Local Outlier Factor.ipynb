```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generating two features X1 and X2 (random data)
X1 = np.random.normal(0, 1, size=1000)
X2 = np.random.normal(0, 1, size=1000)
```

```python
#creating X from X1 and X2 features
X = np.array(list(zip(X1,X2)))
print(X1[0], X2[0], X[0])
```

```python
#visualizing data points
plt.scatter(X[:,0], X[:,1])
```

```python
#importing LocalOutlierFactor from sklearn
from sklearn.neighbors import LocalOutlierFactor
```

```python
#creating instance of LocalOutlierFactor with 50 nearest neightbours
model = LocalOutlierFactor(n_neighbors=50)
```

```python
#fitting model and predecting y_pred on X
y_pred = model.fit_predict(X)
```

```python
#visualizing results
plt.scatter(X[:,0], X[:,1], c=y_pred)
```
