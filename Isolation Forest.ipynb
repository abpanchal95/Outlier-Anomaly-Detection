```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generating two features X1 and X2 (random data)
X1 = np.random.normal(0, 1, size=1000)
X2 = np.random.normal(0, 1, size=1000)
```

```python
#creating X from X1 and X2 features
X = np.array(list(zip(X1,X2)))
print(X1[0], X2[0], X[0])
```

```python
#visualizing data points
plt.scatter(X[:,0], X[:,1])
```

```python
#importing Isolation Forest from sklearn
from sklearn.ensemble import IsolationForest
```

```python
#creating instance of IsolationForest with 100 trees, 50% sampling size for each tree, and 10% samples as anomaly
#max_samples and contamination can be set to auto
model = IsolationForest(n_estimators=100, max_samples=0.5, contamination=0.1)
```

```python
#fitting model and predecting y_pred on X
y_pred = model.fit_predict(X)
```

```python
#visualizing results
plt.scatter(X[:,0], X[:,1], c=y_pred)
```

```python
#All Yellow Points are Normal and Violet Points are detected as Anomaly
#So ratio of yellow points and violets points is 10% as we set contamination to 0.1
```

```python
#here output one means point is normal and not an anomaly
model.predict([[-1, 1]])
```

```python
#here output negative one means point is an anomaly
model.predict([[-3, 1]])
```
