```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generating two features X1 and X2 (random data)
X1 = np.random.normal(0, 1, size=1000)
X2 = np.random.normal(0, 1, size=1000)
```

```python
#creating X from X1 and X2 features
X = np.array(list(zip(X1,X2)))
print(X1[0], X2[0], X[0])
```

```python
#visualizing data points
plt.scatter(X[:,0], X[:,1])
```

```python
#importing EllipticEnvelope from sklearn
from sklearn.covariance import EllipticEnvelope
```

```python
#creating instance of EllipticEnvelope with 0.05 contamination
model = EllipticEnvelope(contamination=0.05)
```

```python
#fitting model and predecting y_pred on X
y_pred = model.fit_predict(X)
```

```python
#visualizing results
plt.scatter(X[:,0], X[:,1], c=y_pred)
```
